```
import numpy as np
import matplotlib.pyplot as plt
```

```
T = 100 #100msec
dt = 0.25 #step size
time = np.arange(0, T+dt, dt)
```

```
#LIF properties
Vm = np.zeros(len(time))
Rm = 2 #in kOhms
Cm = 5 #in microf
tau_m = Rm * Cm
tau_ref = 5 #refactory period
Vth = 1 #in mV
```

```
#simulating the neuron for different inputs I
#part 1
I = np.zeros(len(time))
for i,t in enumerate(time):
    if t < 10:
        I[i] = 0
    elif t > 75:
        I[i] = 0
    else:
        I[i] = 2 #in microA
```

```
plt.plot(time, I)
```

```
#the neuron
def simulate(I):
    t_rest = 0
    for i,t in enumerate(time):
        if t > t_rest:
            Vm[i] = Vm[i-1] + (-Vm[i-1] + I[i-1]*Rm) * dt / tau_m

        if Vm[i] >= Vth:
            t_rest += tau_ref
    return Vm
```

```
Vm = simulate(I)
plt.plot(time, Vm)
```

```
#part 2
for i, t in enumerate(time):
    if t>0 and t<100:
        I[i] = 1.5
    else:
        I[i] = 0

plt.plot(time, I)
```

```
Vm = simulate(I)
plt.plot(time, Vm)
```

```
#part 3
I = 5 * np.sin(np.pi * 0.25 * time)

plt.plot(time, I)
```

```
Vm = simulate(I)
plt.plot(time, Vm)
```

```
#part 4
I = -2*np.sin(0.25*np.pi*time) + 3*np.sin((np.pi/3)*time) + np.cos((np.pi/6)*time)
plt.plot(time, I)
```

```
Vm = simulate(I)
plt.plot(time, Vm)
```

```
#plot u vs i for a constant input
#lets use 20 data points, 1.5microA to 11.5microaA
current = np.zeros((100))
freq = np.zeros((100))
for j in range(100):
    current[j] = 1.5 + 0.1 * j
    I[:] = current[j]
    Vm = simulate(I)
    V_max = np.max(Vm)
    freq[j] = np.sum(Vm == V_max)*10
```

```
plt.plot(freq, current)
```
